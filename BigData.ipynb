```python
import pandas as pd

# Cargar el conjunto de datos
archivo_csv = 'data.csv'
df = pd.read_csv(archivo_csv)

# Anlisis agrupando por 'country' y calculando algunas estadsticas
grupo_por_country = df.groupby('country').agg({
    'name': 'count',        # Nmero de registros por pas
    'phone': 'nunique',     # Nmero de nmeros de telfono nicos por pas
    'email': 'nunique',     # Nmero de direcciones de correo electrnico nicas por pas
    'numberrange': ['min', 'max', 'mean']  # Estadsticas sobre la columna 'numberrange'
})

# Anlisis agrupando por letra inicial del nombre igual a M
df['letra_inicial'] = df['name'].str[0]
grupo_por_letra_M = df[df['letra_inicial'] == 'M'].groupby('letra_inicial').agg({
    'name': 'count',
    'phone': 'nunique',
    'email': 'nunique',
    'numberrange': ['min', 'max', 'mean']
})
df.drop('letra_inicial', axis=1, inplace=True)

# Anlisis agrupando por regin igual a Morelos
grupo_por_morelos = df[df['region'] == 'Morelos'].groupby('region').agg({
    'name': 'count',
    'phone': 'nunique',
    'email': 'nunique',
    'numberrange': ['min', 'max', 'mean']
})

# Muestra los resultados
print("Anlisis agrupado por country:")
print(grupo_por_country)

print("\nAnlisis agrupado por letra inicial igual a M:")
print(grupo_por_letra_M)

print("\nAnlisis agrupado por regin igual a Morelos:")
print(grupo_por_morelos)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
archivo_csv = 'data.csv'
df = pd.read_csv(archivo_csv)

# Anlisis agrupando por 'country' y calculando algunas estadsticas
grupo_por_country = df.groupby('country').agg({
    'name': 'count',
    'phone': 'nunique',
    'email': 'nunique',
    'numberrange': ['min', 'max', 'mean']
})

# Visualizacin por Country en el eje X y Region en el eje Y
# Aqu asumimos que 'region' es una columna en tu conjunto de datos
# Puedes ajustar segn la estructura real de tus datos
x = grupo_por_country.index
y = df.groupby('country')['region'].nunique()

# Configurar y mostrar el grfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Unique Regions')
plt.title('Number of Unique Regions per Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
archivo_csv = 'data.csv'
df = pd.read_csv(archivo_csv)

# 1. Exploracin de Datos
# Crear un DataFrame con informacin sobre cada columna y el tipo de valor que contienen
info_columnas = pd.DataFrame({
    'Columna': df.columns,
    'Tipo de Dato': df.dtypes.values,
    'No Nulos': df.count().values,
    'No nicos': df.nunique().values
})

# 2. Transformaciones y Filtrado
# Filtrar datos para mostrar los primeros 500 datos
primeros_500 = df.head(500)

# Filtrar datos por country donde country sea igual a Mxico
datos_mexico = df[df['country'] == 'Mexico']

# Filtrar datos por country donde country sea igual a Espaa
datos_espana = df[df['country'] == 'Spain']

# Filtrar datos para que las columnas estn ordenadas por nombre de A a Z
df_ordenado = df.sort_values(by='name', ascending=True)

# Filtrar datos para email donde el email contenga @outlook.com
datos_outlook = df[df['email'].str.contains('@outlook.com')]

# 3. Anlisis de Datos
# Anlisis agrupado por 'country' y calculando algunas estadsticas
grupo_por_country = df.groupby('country').agg({
    'name': 'count',
    'phone': 'nunique',
    'email': 'nunique',
    'numberrange': ['min', 'max', 'mean']
})

# 4. Visualizacin de Datos
# Visualizacin por Country en el eje X y Region en el eje Y
x = grupo_por_country.index
y = df.groupby('country')['region'].nunique()

# Configurar y mostrar el grfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Unique Regions')
plt.title('Number of Unique Regions per Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('grafico.png')  # Guardar el grfico como imagen

# Exportar los resultados a archivos CSV
info_columnas.to_csv('info_columnas.csv', index=False)
primeros_500.to_csv('primeros_500.csv', index=False)
datos_mexico.to_csv('datos_mexico.csv', index=False)
datos_espana.to_csv('datos_espana.csv', index=False)
df_ordenado.to_csv('df_ordenado.csv', index=False)
datos_outlook.to_csv('datos_outlook.csv', index=False)
grupo_por_country.to_csv('grupo_por_country.csv')

# Devolver informacin de cada columna y el tipo de valor que contienen
info_columnas = df.dtypes.reset_index()
info_columnas.columns = ['Columna', 'Tipo de Dato']

# Guardar informacin de columnas en un archivo CSV
info_columnas.to_csv('info_columnas.csv', index=False)
```
