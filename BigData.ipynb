```python
import pandas as pd

# Cargar el conjunto de datos
archivo_csv = 'data.csv'
df = pd.read_csv(archivo_csv)

# Análisis agrupando por 'country' y calculando algunas estadísticas
grupo_por_country = df.groupby('country').agg({
    'name': 'count',        # Número de registros por país
    'phone': 'nunique',     # Número de números de teléfono únicos por país
    'email': 'nunique',     # Número de direcciones de correo electrónico únicas por país
    'numberrange': ['min', 'max', 'mean']  # Estadísticas sobre la columna 'numberrange'
})

# Análisis agrupando por letra inicial del nombre igual a M
df['letra_inicial'] = df['name'].str[0]
grupo_por_letra_M = df[df['letra_inicial'] == 'M'].groupby('letra_inicial').agg({
    'name': 'count',
    'phone': 'nunique',
    'email': 'nunique',
    'numberrange': ['min', 'max', 'mean']
})
df.drop('letra_inicial', axis=1, inplace=True)

# Análisis agrupando por región igual a Morelos
grupo_por_morelos = df[df['region'] == 'Morelos'].groupby('region').agg({
    'name': 'count',
    'phone': 'nunique',
    'email': 'nunique',
    'numberrange': ['min', 'max', 'mean']
})

# Muestra los resultados
print("Análisis agrupado por country:")
print(grupo_por_country)

print("\nAnálisis agrupado por letra inicial igual a M:")
print(grupo_por_letra_M)

print("\nAnálisis agrupado por región igual a Morelos:")
print(grupo_por_morelos)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
archivo_csv = 'data.csv'
df = pd.read_csv(archivo_csv)

# Análisis agrupando por 'country' y calculando algunas estadísticas
grupo_por_country = df.groupby('country').agg({
    'name': 'count',
    'phone': 'nunique',
    'email': 'nunique',
    'numberrange': ['min', 'max', 'mean']
})

# Visualización por Country en el eje X y Region en el eje Y
# Aquí asumimos que 'region' es una columna en tu conjunto de datos
# Puedes ajustar según la estructura real de tus datos
x = grupo_por_country.index
y = df.groupby('country')['region'].nunique()

# Configurar y mostrar el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Unique Regions')
plt.title('Number of Unique Regions per Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
archivo_csv = 'data.csv'
df = pd.read_csv(archivo_csv)

# 1. Exploración de Datos
# Crear un DataFrame con información sobre cada columna y el tipo de valor que contienen
info_columnas = pd.DataFrame({
    'Columna': df.columns,
    'Tipo de Dato': df.dtypes.values,
    'No Nulos': df.count().values,
    'No Únicos': df.nunique().values
})

# 2. Transformaciones y Filtrado
# Filtrar datos para mostrar los primeros 500 datos
primeros_500 = df.head(500)

# Filtrar datos por country donde country sea igual a México
datos_mexico = df[df['country'] == 'Mexico']

# Filtrar datos por country donde country sea igual a España
datos_espana = df[df['country'] == 'Spain']

# Filtrar datos para que las columnas estén ordenadas por nombre de A a Z
df_ordenado = df.sort_values(by='name', ascending=True)

# Filtrar datos para email donde el email contenga @outlook.com
datos_outlook = df[df['email'].str.contains('@outlook.com')]

# 3. Análisis de Datos
# Análisis agrupado por 'country' y calculando algunas estadísticas
grupo_por_country = df.groupby('country').agg({
    'name': 'count',
    'phone': 'nunique',
    'email': 'nunique',
    'numberrange': ['min', 'max', 'mean']
})

# 4. Visualización de Datos
# Visualización por Country en el eje X y Region en el eje Y
x = grupo_por_country.index
y = df.groupby('country')['region'].nunique()

# Configurar y mostrar el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Unique Regions')
plt.title('Number of Unique Regions per Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('grafico.png')  # Guardar el gráfico como imagen

# Exportar los resultados a archivos CSV
info_columnas.to_csv('info_columnas.csv', index=False)
primeros_500.to_csv('primeros_500.csv', index=False)
datos_mexico.to_csv('datos_mexico.csv', index=False)
datos_espana.to_csv('datos_espana.csv', index=False)
df_ordenado.to_csv('df_ordenado.csv', index=False)
datos_outlook.to_csv('datos_outlook.csv', index=False)
grupo_por_country.to_csv('grupo_por_country.csv')

# Devolver información de cada columna y el tipo de valor que contienen
info_columnas = df.dtypes.reset_index()
info_columnas.columns = ['Columna', 'Tipo de Dato']

# Guardar información de columnas en un archivo CSV
info_columnas.to_csv('info_columnas.csv', index=False)
```
